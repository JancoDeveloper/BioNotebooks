## Tuplas (Datos inmutables)

Las tuplas son m谩s r谩pidas que las listas. Si define un conjunto constante de valores y todo lo que va a hacer es iterar sobre ellos, use una tupla en lugar de una lista.

Los items de una tupla pueden ser objetos de cualquier tipo.

Para especificar una tupla, lo hacemos con los elementos separados por comas dentro de par茅ntesis.

Una tupla con 煤nicamente dos elementos es denominada par.

Para crear una tupla con un 煤nico elemento (singleton), se a帽ade una coma al final de la expresi贸n.

Para definir una tupla vac铆a, se emplean unos par茅ntesis vac铆os.

```python
tupla = ('cadena de texto', 15, 2.8, 'otro dato', 25,25)
print(tupla)
print(type(tupla))
print(tupla[1:4])
print('cuantas veces hay un elemento=',tupla.count(25))
print('dice en que posicion esta el elemento=',tupla.index(15))
tupla[0]=2# dado que es una TUPLA no se puede agregar ni borrar informaci贸n.
```

## Diccionarios

Los diccionarios de Python son una lista de consulta de t茅rminos de los cuales se proporcionan valores asociados.
En Python, un diccionario es una colecci贸n no-ordenada de valores que son accedidos a traves de una clave. Es decir, en lugar de acceder a la informaci贸n mediante el 铆ndice num茅rico, como es el caso de las listas y tuplas, es posible acceder a los valores a trav茅s de sus claves, que pueden ser de diversos tipo.
Las claves son 煤nicas dentro de un diccionario, es decir que no puede haber un diccionario que tenga dos veces la misma clave, si se asigna un valor a una clave ya existente, se reemplaza el valor anterior.
No hay una forma directa de acceder a una clave a trav茅s de su valor, y nada impide que un mismo valor se encuentre asignado a distintas claves.
La informacion almacenada en los diccionarios, no tiene un orden particular. Ni por clave ni por valor, ni tampoco por el orden en que han sido agregados al diccionario.
Cualquier variable de tipo inmutable, puede ser clave de un diccionario: cadenas, enteros, tuplas (con valores inmutables en sus miembros), etc. No hay restricciones para los valores que el diccionario puede contener, cualquier tipo puede ser el valor: listas, cadenas, tuplas, otros diccionarios, objetos, etc.

```python
# Definir una variable diccionario
futbolistas = dict()

futbolistas = {
    1: "Casillas", 3: "Piqu茅",
    5: "Puyol", 6: "Iniesta",
    7: "Villa", 8: "Xavi Hern谩ndez",
    9: "Torres", 11: "Capdevila",
    14: "Xavi Alonso", 15: "Ramos",
    16: "Busquets"
}

# Recorrer el diccionario, imprimiendo clave - valor
print ("Vemos que los elementos no van \"ordenados\":")
for k, v in futbolistas.items():
    print ("{} --> {}".format(k,v))

# N潞 de elementos que tiene un diccionario
numElemen = len(futbolistas)
print ("\nEl n煤mero de futbolistas es de {}".format(numElemen))

# Imprimir las claves que tiene un diccionario
keys = futbolistas.keys();
print ("\nLas claves de nuestro diccionario son : {}".format(keys))

# Imprimir los valores que tiene un diccionario
values = futbolistas.values();
print ("\nLos valores de nuestro diccionario son : {}".format(values))

# Obtener el valor de un elemento dada su clave
elem = futbolistas.get(6)
print ("\nEl futbolista con clave 6 es {}".format(elem))

# Insertamos un elemento en el diccionario
## si la clave ya existe, el elemento NO se inserta
futbolistas.setdefault(10, 'Cesc')
print ("\nInsertamos el elemento con clave 10 y valor Cesc")
print ("Ahora el diccionario queda : {}".format(futbolistas))

# A帽adimos, de otro modo, un elemento al diccionario
## si la clave ya existe, el valor se cambia por este nuevo
futbolistas[22] = 'Navas'
print ("\nA帽adimos un nuevo elemento, ahora el diccionario queda: ")
print (futbolistas)

# Eliminamos un elemento del diccionario dada su clave
futbolistas.pop(22)
print ("\nEliminamos el elemento con clave 22")
print ("Ahora el diccionario queda: {}".format(futbolistas))

# Hacemos una copia del diccionario
futbolistas_copia = futbolistas.copy()
print ("\nLa copia del diccionario tiene los valores:")
print (futbolistas_copia)

# Borramos los elementos de un diccionario
futbolistas_copia.clear()
print ("\nVaciamos el diccionario nuevo creado, ahora los valores en el: {}".format(futbolistas_copia))

# Creamos un diccionario a partir de una lista de claves
keys = ['nombre', 'apellidos', 'edad']
datos_usuario = dict.fromkeys(keys, 'null')
print ("\nCreamos un diccionario a partir de la lista de claves:")
print (keys)
print ("y con valor 'null'")
print ("As铆 el diccionario queda: {}".format(datos_usuario))

# Comprobamos si existe o no una clave en un diccionario
if 2 in futbolistas:
    print ("\nEl futbolista con la clave 2 existe en el diccionario.")
else:
    print ("\nEl futbolista con la clave 2 NO existe en el diccionario.")

if 8 in futbolistas:
    print ("\nEl futbolista con la clave 8 existe en el diccionario.")
else:
    print ("\nEl futbolista con la clave 8 NO existe en el diccionario.")

# Devolvemos los elementos del diccionario en una tupla
tupla = futbolistas.items()
print ("\nEl diccionario convertido en tupla queda as铆:")
print (tupla)

# Unimos dos diccionarios existentes
suplentes = {
    4:'Marchena', 12:'Valdes', 13:'Mata',
    17:'Arbeloa', 19:'Llorente', 20:'Javi Martinez',
    21:'Silva', 23:'Reina'
}

print ("\nUnimos el diccionario: ")
print (futbolistas)
futbolistas.update(suplentes) # Aqu铆 hacemos la uni贸n de los diccionarios
print ("con el diccionario:")
print (suplentes)
print ("siendo el resultado:")
print (futbolistas)
```

```python
# funci贸n para determinar el cociente
def funcionCociente(x,w):
  q=0
  r=x
  while r>=w:
    q=q+1
    r=r-w
  return q 
```

```python
# funci贸n para determinar el residuo
def funcionResiduo(x,w):
  q=0
  r=x
  while r>=w:
    q=q+1
    r=r-w
  return r 
```

```python
# funci贸n para determinar si un numero es par
def funcionPar(u):
  if funcionResiduo(u,2)==0:
    par=True
  else:
    par=False
  return par    
```

```python
def funcionMultiplicar(x,w):
  z=0
  u=x
  v=w 
  while u!=0:
    if not funcionPar(u):
      z=z+v
    u=funcionCociente(u,2)
    v=2*v
  return z
```

```python
def funcionDividirTabla(x,y):
  q=0
  w=y
  r=x 
  paso = 1
  print( "Paso" + "   " + "Q" + "   " + "W" + "   " + "R")
  print( "-----------------------------------------------------")
  print( str(paso) + "   " + str(q) + "   " + str(w) + "   " + str(r))
  while w<=r:
    w = 2*w
    print(str(paso) + "   " + str(q) + "   " + str(w) + "   " + str(r))
  while w!=y:
    q=2*q
    w=w/2
    if w<=r:
      q=q+1
      r=r-w
    print(str(paso) + "   " + str(q) + "   " + str(w) + "   " + str(r))
  paso=paso+1
  print(str(paso) + "   " + str(q) + "   " + str(w) + "   " + str(r))
  return q
```

```python
x=int(input('Escribir un numero par dividir: '))
y=int(input('Escribir otro numero para dividir: '))
print('El resultado de la division es: ', funcionDividirTabla(x,y))
```

## Ejecricio 3

Construir una matriz aleatoria de cuatro letras (A,T,C,G) de tama帽o 100x1000 luego cambiar las letras por las siguientes codificaciones:

Codificaci贸n 1: -2,-1,1,2

Codificaci贸n 2: 0,1,2,3

Codifiaci贸n 3: 00,01,10,11


```python
import numpy as np
k = np.random.choice(('A','T','C','G'),size=(100,1000))
cod1 = np.zeros(k.shape, dtype=np.int8)
cod2 = np.zeros(k.shape, dtype=np.int8)
cod3 = np.empty(k.shape, dtype="S2")

dic1 = dict()
dic1 = {
    'A': -2,
    'T': -1,     
    'C': 1, 
    'G': 2
}

dic2 = dict()
dic2 = {
    'A': 0,
    'T': 1,     
    'C': 2, 
    'G': 3
}

dic3 = dict()
dic3 = {
    'A': "00",
    'T': "01",     
    'C': "10", 
    'G': "11"
}

k.shape[1]
k.shape[0]

for j in range(k.shape[1]):
  for i in range(k.shape[0]):
    cod1[i,j] = dic1.get(k[i,j])
    cod2[i,j] = dic2.get(k[i,j])
    cod3[i,j] = dic3.get(k[i,j])

print(k[0,0])
print(cod1[0,0])
print(cod2[0,0])
print(cod3[0,0].decode("utf-8"))

print(k)
print(cod1)
print(cod2)
print(cod3)
```
